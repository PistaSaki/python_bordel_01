```python
import tensorflow as tf
import numpy as np
from numpy import random as rnd

import tensorflow.keras.layers as kl
```

```python
x = rnd.rand(1, 5, 2).astype(np.float32)
```

```python
l = kl.LSTM(2)
out1 = l(x)
out2 = l(x)
np.allclose(out1, out2)
```

```python
l = kl.LSTM(2, stateful=True)
out1 = l(x)
out2 = l(x)
np.allclose(out1, out2)
```

### Is state updated in trainig?

```python
def get_loss(x):
    return tf.reduce_sum(l(x) ** 2)
```

```python
opt = tf.optimizers.Adam()
```

```python
with tf.GradientTape() as tape:
    loss = get_loss(x)

grad = tape.gradient(loss, l.trainable_weights)
```
