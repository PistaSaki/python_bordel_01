```python
import os
from pathlib import Path
import json
import numpy as np
import matplotlib.pyplot as pl
from typing import List, Dict, Union
%matplotlib inline
```

```python
in_dir = Path("d:/tmp")
out_dir = in_dir
```

```python
def _plot(model_name, dif_lines, probas_draw, probas_1st, probas_2nd, sum_line, time_to_end, current_dif, ball = None ) -> None:
    pl.plot(dif_lines, probas_draw, label="draw")
    pl.plot(dif_lines, probas_1st, label="1st wins")
    pl.plot(dif_lines, probas_2nd, label="2nd wins")
    pl.legend()
    pl.xlabel("asian handicap line")
    pl.ylabel("proba")
    pl.suptitle(f"{model_name} with ou_line = {sum_line}, time_to_end = {time_to_end}, ball = {ball}, current_dif = {current_dif}")
 
```

```python
def _get_out_path(in_path, out_dir) -> Path:
    out_dir = Path(out_dir)
    p = Path(in_path)
    return out_dir / p.stem 
```

```python
for in_path in in_dir.glob("**/*.json"):
    with open(in_path, "rt") as in_file:
        dic = json.load(in_file)
        
        
    pl.figure(figsize=[10, 8])
    _plot(**dic)
    pl.savefig(_get_out_path(in_path=in_path, out_dir=out_dir))
```
