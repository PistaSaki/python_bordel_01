```python
import scipy.stats as st
import matplotlib.pyplot as pl
import numpy as np
```

Assume symmetric prior $ p \sim B(\alpha, \alpha)$.

```python
alpha = 1
post_a = st.beta(a = alpha + 2, b = alpha + 1)
post_b = st.beta(a = alpha + 1, b = alpha + 0)

xxx = np.linspace(0, 1, 100)
pl.plot(xxx, post_a.pdf(xxx), label = "a")
pl.plot(xxx, post_b.pdf(xxx), label = "b")
pl.gca().set_title(f"posterior pdfs for $\\alpha$ = {alpha}")
pl.axvline(1/2, linestyle = "--", color = "grey")
pl.legend()

print(
    f"Posterior probas that p > 1/2 are "
    f"a) {1 - post_a.cdf(1/2)}, b) {1 - post_b.cdf(1/2)}"
)
```
