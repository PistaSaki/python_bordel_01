```python
import numpy as np
from tensorflow.keras import backend as K
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
```

```python
X, Y = np.mgrid[0:1:10j, 0:1:10j]
```

```python
K.cast_to_floatx(1)
```

```python
def f(x, y):
    x, y = [tf.constant(v, K.floatx()) for v in [x, y]]
    return K.binary_crossentropy(x, y).numpy()
```

```python
Z = f(X, Y)
```

```python
plt.pcolormesh(Z)
```

```python
for p in [0.1, 0.4, 0.5, 0.6, 0.8, 0.9]:
    xxx = np.linspace(0, 1, 20)
    yyy = np.exp( -f(p, xxx))
    plt.plot(xxx, yyy, label=f"{p}")

plt.legend()
```

```python
for p in [0.1, 0.4, 0.5, 0.6, 0.8, 0.9, 0.95, 0.99]:
    xxx = np.linspace(0, 1, 20)
    yyy = np.exp( -f(p, xxx))
    yyy /= yyy.sum()
    plt.plot(xxx, yyy, label=f"{p}")

plt.legend()
```

```python
ppp = np.linspace(0, 1, 20)
xxx = np.linspace(0, 1, 20)
zzz = np.exp(-f(ppp[:, None], xxx[None, :]))
plt.plot(ppp, zzz.sum(axis=1))
```
